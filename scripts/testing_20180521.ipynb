# Composite KEGGeractome

```python
from glob import glob
from os.path import basename

paths = glob("../data/human_KGML/*.xml")

graphs = {}
for path in paths: 
    try: 
        KEGG_pathway = KEGGeractome(path)
        graphs[KEGG_pathway.name] = KEGG_pathway
    except: 
        pass
```

```python
# does not include node attributes yet
all_nodes = list(set(flatten([keggie.graph.nodes() for keggie in graphs.values()])))
len(all_nodes)
```

```python
# 
all_edges_data = flatten([keggie.graph.edges(data=True) for keggie in graphs.values()])
edge_node_symbols_df = pd.DataFrame([[source, target] for source,target,_ in all_edges_data], columns=["source", "target"])
edge_attributes_df = pd.DataFrame([attribs for _,_,attribs in all_edges_data])
all_edges_df = pd.concat([edge_node_symbols_df,edge_attributes_df], axis=1)

all_edges_df.head()
```

```python
all_edges_df.drop_duplicates(inplace=True)
all_edges_df.sort_values(by=["source", "target", "oriented", "type"], 
                         ascending=[True, True, True, True], 
                         inplace=True)
```
